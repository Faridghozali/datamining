Mengubah type dan rating
* jika movie/PG-13,jadi angka 1
* jika TV Show/TV-MA jadi angka 0

```python
# type code
type_movie = "Movie"
type_Tv = "TV Show"

# binary value
value_of_one = 1
value_of_zero = 0

def change_code_embark_to_biner(type):
    return value_of_one if type == type_movie else value_of_zero
```

```python
# rating code
rating_movie = "PG-13"
rating_Tv = "TV-MA"

# binary value
value_of_one = 1
value_of_zero = 0

def change_code_rating_to_biner(rating):
    return value_of_one if rating == rating_movie else value_of_zero
```

```python
# Update all values of 'type' series
netflix_titles["type"] = netflix_titles["type"].apply(change_code_embark_to_biner)
netflix_titles["rating"] = netflix_titles["rating"].apply(change_code_rating_to_biner)
```

```python
# CONSTAN VARIABLE
DECREMENT_BY_ONE = 1
INCREMENT_BY_ONE = 1

CONTINGENCY_TABLE_VALUE = {
    "q" : (1,1),
    "r" : (1,0),
    "s" : (0,1),
    "t" : (0,0),
}
```

```python
def get_series(df, idx, series):
    return netflix_titles.loc[(idx), series]
```

```python
def get_dissimilarity_dataset(netflix_titles, series_index = [], series = []):
    first_series = get_series(netflix_titles, series_index[0], series)
    second_series = get_series(netflix_titles, series_index[1], series)
    dataset = pd.concat([first_series,second_series],axis=1)
    return dataset.T
```

```python
def count_contingency_value(netflix_titles, start_index = 0, last_index = 1):

    CONTINGENCY_VALUE = {
        "q" : 0,
        "r" : 0,
        "s" : 0,
        "t" : 0,
    }

    column_range = netflix_titles.shape[1]

    for column in range(column_range):
        for value in CONTINGENCY_TABLE_VALUE:
            item = list((tuple(netflix_titles.loc[(start_index):(last_index), netflix_titles.columns[column]]) == CONTINGENCY_TABLE_VALUE[value], value))
            if item[0] == True:
                if item[1] == "q":
                    CONTINGENCY_VALUE["q"] += 1
                if item[1] == "r":
                    CONTINGENCY_VALUE["r"] += 1
                if item[1] == "s":
                    CONTINGENCY_VALUE["s"] += 1
                if item[1] == "t":
                    CONTINGENCY_VALUE["t"] += 1

    return CONTINGENCY_VALUE
```

```python
# d(1,2)
df_1_2 = get_dissimilarity_dataset(netflix_titles, [1,2], ["type", "rating"])
```

```python
c_d_1_2 = count_contingency_value(df_1_2, 1, 2)
```
