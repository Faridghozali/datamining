**Definisi Binning berbasis entropi**

1.   Metode untuk mengelompokkan data *numeric* menjadi *categorical*
2.   Pengelompokkan dengan mencari jumlah pembagi yang terabaik
3.   Hasil pengelompokkan terbaik adalah dengan *entropy gain* yang paling besar


*  Rumus Entropy

$$
\begin{align*}
\displaystyle Entropy(S) &= \sum_{i=0}^{k} -pi \ log_{2} \ pi \\
\end{align*}
$$

**Fungsi menghitung banyaknya data setiap kategori**

````python
# hitung setiap nilai berdasarkan kategori yang diberikan
def countEveryCategory(data_column, labels, column, category):
    grup = data_column.groupby(category).count()
    amount_of_every_category = []
    for label in labels:
        amount_of_every_category.append(grup.loc[label, column])
    return amount_of_every_category
````

**Fungsi menghitung banyaknya data setiap kategori berdasarkan nilai pembagi atau *split value* yang diberikan**

````python
# membagi data yang diberikan berdasarkan kategori dan nilai split
def split(split_value, data_column, labels, col, category):
    less_grup = data_column[data_column[col] < split_value]
    greater_grup = data_column[data_column[col] >= split_value]
    
    length_less_grup = countEveryCategory(less_grup, labels, col, category)
    length_greater_grup = countEveryCategory(greater_grup, labels, col, category)
    
    return (length_less_grup, length_greater_grup)
````

**Fungsi menghitung perbedaan *gain* antara entropi inisial dengan entropi nilai pembagi yang diberikan**




---
$$Gain(E_{new}) = (E_{initial}) \ - (E_{new})$$
---



````python
# membagi data yang diberikan berdasarkan kategori dan nilai split
def count_gain(inisial_entropy, new_entropy):
  return inisial_entropy - new_entropy
````

**Fungsi menghitung entropy**

````python
# menghitung entropi
def count_entropy(data_tujuan):
    all_prob = []
    for prob in data_tujuan:
        if (prob/sum(data_tujuan) != 0):
            all_prob.append(prob/sum(data_tujuan) * log2(prob/sum(data_tujuan)))
        else:
            all_prob.append(0)
    return -(sum(all_prob))
````

**Fungsi menghitung Info entropy berdasarkan nilai pembagi atau *split value* yang diberikan**


---



$$ Info_A(D) = \frac{|D_1|}{|D|} Entropy (D_1)\frac{|D_2|}{|D|}+ Entropy (D_2)$$

````python
# hitung entropi untuk nilai split yang diberikan
def info(d, data_tujuan):
    temp = []
    for value in d:
        temp.append((sum(value) / data_tujuan.shape[0]) * count_entropy(value))
    return sum(temp)
````
